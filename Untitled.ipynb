```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/AirlineETL')
connection = engine.connect()
```

```python
airport_info = pd.read_csv("Data/airports.csv")
countries_info = pd.read_csv("Data/countries.csv")
route_info = pd.read_csv("Data/routes.csv")
```

```python
route_df = route_info.drop(columns=[" codeshare", " stops"])
route_df.rename(columns={
    'airline ID': "airline_id",
    ' source airport': "source_airport",
    ' source airport id': "source_airport_id",
    ' destination apirport': "destination_airport",
    ' destination airport id': "destination_airport_id",
    ' equipment': "equipment"
}, inplace=True)


# route_df = route_df[route_df["source_airport"] != "IKE", "IOT", "JJA", "JUU", "MLH", "QFN", "QJH", 
#                     "QUV", "QUW", "SHP", "SXF", "UPK", "XEQ", "ZBL"]

route_df
```

```python
# Use pandas to load csv converted DataFrame into database
route_df.to_sql(name='airline_routes', con=engine, if_exists='append', index=False)

# Use pandas to load json converted DataFrame into database
#new_customer_location_df.to_sql(name='customer_location', con=engine, if_exists='append', index=False)
```

```python
airport_df = airport_info.drop(columns=["ident", "elevation_ft", "continent", "iso_region",
                                        "municipality", "scheduled_service", "gps_code", "local_code",
                                       "home_link", "wikipedia_link", "keywords"])

airport_df.rename(columns={
    "type": "ap_type",
    "name": "ap_name",
    "latitude_deg": "lat",
    "longitude_deg": "lng",
    "iso_country": "country_code",
}, inplace=True)

no_null_airport_df = airport_df.dropna(how='any')

no_null_airport_df
```

```python
# Use pandas to load csv converted DataFrame into database
no_null_airport_df.to_sql(name='airports', con=engine, if_exists='append', index=False)

# Use pandas to load json converted DataFrame into database
#new_customer_location_df.to_sql(name='customer_location', con=engine, if_exists='append', index=False)
```

```python
countries_df = countries_info.drop(columns=["wikipedia_link", "keywords", "continent"])

#countries_df.rename(columns={
#    "continent": "continent_id"
#}, inplace=True)

#countries_df.dropna(how='any',inplace=True)
countries_df
```
