```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
# read in csv files
engine = create_engine('postgresql://postgres:postgres@localhost:5432/AirlineETL')
connection = engine.connect()
```

```python
# read in csv files
airport_info = pd.read_csv("Data/airports.csv")
countries_info = pd.read_csv("Data/countries.csv")
route_info = pd.read_csv("Data/routes.csv")
```

```python
countries_df = countries_info.drop(columns=["wikipedia_link", "keywords", "continent"])

countries_df
```

```python
countries_df["code"].value_counts

for index,row in countries_df.iterrows():
    if row[2] == "Namibia":
        countries_df.loc[index,"code"] = "N"
        
countries_df[countries_df["code"] == 'N']
```

```python
# Use pandas to load csv converted DataFrame into database
countries_df.to_sql(name='countries', con=engine, if_exists='append', index=False)
```

```python
airport_df = airport_info.drop(columns=["ident", "elevation_ft", "continent", "iso_region",
                                        "municipality", "scheduled_service", "gps_code", "local_code",
                                       "home_link", "wikipedia_link", "keywords"])

airport_df.rename(columns={
    "type": "ap_type",
    "name": "ap_name",
    "latitude_deg": "lat",
    "longitude_deg": "lng",
    "iso_country": "country_code",
}, inplace=True)

no_null_airport_df = airport_df.dropna(how='any')

no_null_airport_df = no_null_airport_df.loc[(no_null_airport_df["ap_type"] != "closed") & 
                                            (no_null_airport_df["iata_code"] != "PHM") &
                                            (no_null_airport_df["iata_code"] != "ABJ") &
                                            (no_null_airport_df["iata_code"] != "GGC") &
                                            (no_null_airport_df["iata_code"] != "CTY") &
                                            (no_null_airport_df["iata_code"] != "MRE") &
                                            (no_null_airport_df["iata_code"] != "KMM") &
                                            (no_null_airport_df["iata_code"] != "PRM") &
                                            (no_null_airport_df["iata_code"] != "DDU") &
                                            (no_null_airport_df["iata_code"] != "KQA") &
                                            (no_null_airport_df["iata_code"] != "PCO") &
                                            (no_null_airport_df["iata_code"] != "NWT")]

no_null_airport_df.duplicated(keep = 'last')
no_null_airport_df = no_null_airport_df.loc[no_null_airport_df["iata_code"].duplicated() != True]
clean_airport_df = no_null_airport_df
        
clean_airport_df
```

```python
# Use pandas to load csv converted DataFrame into database
clean_airport_df.to_sql(name='airports', con=engine, if_exists='append', index=False)
```

```python
# create dataframe and drop unnecessary columns
route_df = route_info.drop(columns=[" codeshare", " stops"])

# rename columns
route_df.rename(columns={
    'airline ID': "airline_id",
    ' source airport': "source_airport",
    ' source airport id': "source_airport_id",
    ' destination apirport': "destination_airport",
    ' destination airport id': "destination_airport_id",
    ' equipment': "equipment"
}, inplace=True)

route_df
```

```python
route_df = route_df.loc[(route_df["airline_id"] != 'N') &
                        (route_df["source_airport"] != 'N') &
                        (route_df["source_airport"] != 'ABJ') &
                        (route_df["source_airport"] != 'TXL') &
                        (route_df["source_airport"] != 'ETH') &
                        (route_df["source_airport"] != 'SXF') &
                        (route_df["source_airport"] != 'RTW') &
                        (route_df["source_airport"] != 'MRE') &
                        (route_df["source_airport"] != 'WDH') &
                        (route_df["source_airport"] != 'ISN') &
                        (route_df["source_airport"] != 'DOK') &
                        (route_df["source_airport"] != 'SHP') &
                        (route_df["source_airport"] != 'KCG') &
                        (route_df["source_airport"] != 'KQA') &
                        (route_df["source_airport"] != 'SDV') &
                        (route_df["source_airport"] != 'WVB') &
                        (route_df["source_airport"] != 'ERS') &
                        (route_df["source_airport"] != 'LUD') &
                        (route_df["source_airport"] != 'MPA') &
                        (route_df["source_airport"] != 'NVK') &
                        (route_df["destination_airport"] != 'N') &
                        (route_df["destination_airport"] != 'ABJ') &
                        (route_df["destination_airport"] != 'TXL') &
                        (route_df["destination_airport"] != 'ETH') &
                        (route_df["destination_airport"] != 'SXF') &
                        (route_df["destination_airport"] != 'RTW') &
                        (route_df["destination_airport"] != 'MRE') &
                        (route_df["destination_airport"] != 'WDH') &
                        (route_df["destination_airport"] != 'ISN') &
                        (route_df["destination_airport"] != 'DOK') &
                        (route_df["destination_airport"] != 'SHP') &
                        (route_df["destination_airport"] != 'KCG') &
                        (route_df["destination_airport"] != 'KQA') &
                        (route_df["destination_airport"] != 'SDV') &
                        (route_df["destination_airport"] != 'WVB') &
                        (route_df["destination_airport"] != 'ERS') &
                        (route_df["destination_airport"] != 'LUD') &
                        (route_df["destination_airport"] != 'MPA') &
                        (route_df["destination_airport"] != 'NVK') &
                        (route_df["destination_airport_id"] != 'N') &
                        (route_df["source_airport_id"] != 'N')]

route_df = route_df.dropna(how='any')

route_df
```

```python
# Use pandas to load csv converted DataFrame into database
route_df.to_sql(name='airline_routes', con=engine, if_exists='append', index=False)
```
